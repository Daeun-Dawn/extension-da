## JOIN - INNER JOIN, LEFT JOIN, SELF JOIN

```python
# inner join
# Customers c <- Orders o 
# if c.CustomerID == o.CustomerID
# (INNER) JOIN
query = """
    SELECT c.ContactName, o.OrderID
    FROM Customers c
    JOIN
        Orders o
        ON c.CustomerID = o.CustomerID
    ;
"""

pd.read_sql(query, db).head()
```

```python
# left join
query = """
    SELECT c.ContactName, o.OrderID
    FROM Customers c
    LEFT JOIN 
        Orders o
        ON c.CustomerID = o.CustomerID
    ;
"""

pd.read_sql(query, db).head()
```

```python
#self join
query = """
    SELECT 
        CA.CustomerName "CustomerName1",
        CB.CustomerName "CustomerName2",
        CA.City
        
    FROM Customers CA, Customers CB
    WHERE
        CA.CustomerID <> CB.CustomerID
        AND CA.City = CB.City
    ;
"""

pd.read_sql(query, db).head()
```

## Cohort Analysis with GA

```python
db = lite.connect('../data/1804newuser.db')
```

```python
query = """
    SELECT *
    FROM sqlite_master
    WHERE
        type='table'
    ;
"""

pd.read_sql(query, db)
```

```python
query = """
    SELECT 
        n.date, 
        n.user, 
        round(d0.user *100.0 / n.user, 2) "day0",
        round(d1.user *100.0 / n.user, 2) "day1",
        round(d2.user *100.0 / n.user, 2) "day2",
        round(d3.user *100.0 / n.user, 2) "day3",
        round(d4.user *100.0 / n.user, 2) "day4",
        round(d5.user *100.0 / n.user, 2) "day5"
    FROM NewUser n
    JOIN
        day0 d0
        ON n.date = d0.date
    JOIN
        day1 d1
        ON n.date = d1.date
    JOIN
        day2 d2
        ON n.date = d2.date
    JOIN
        day3 d3
        ON n.date = d3.date
    JOIN
        day4 d4
        ON n.date = d4.date
    JOIN
        day5 d5
        ON n.date = d5.date
    ;
"""

pd.read_sql(query, db)
```
